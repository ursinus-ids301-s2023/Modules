A loop is a structure that we use to repeat code with minor variations many times over again until some condition is satisfied

Arithmetic series

```python
for i in range(10): # Creates a variable called i which starts at 0, increments by 1
    # and then stops right before 10
    print(i, end=' ')
```

```python
def arithmetic_series(N):
    """
    Sum up the numbers from 1 + 2 + 3 + ... + N
    
    Parameters
    ----------
    N: int
        Last number to add
    """
    result = 0 # This variable will "accumulate" the sum of all numbers
    for i in range(N):
        result = result + (i+1)
    return result

print(1+2+3+4+5+6, arithmetic_series(6))
```

```python
arithmetic_series(100)
```

```python
def arithmetic_series_fast(N):
    return N*(N+1)//2
arithmetic_series_fast(6)
```

Square wave approximation

```python
N = 1000
t = np.linspace(0, 1, N)
y = np.zeros(N)
y = y + np.sin(2*np.pi*t)
y = y + np.sin(2*np.pi*3*t)/3
y = y + np.sin(2*np.pi*5*t)/5
y = y + np.sin(2*np.pi*7*t)/7
y = y + np.sin(2*np.pi*9*t)/9
plt.plot(y)
```

```python
N = 1000
t = np.linspace(0, 1, N)
y = np.zeros(N)
n_harmonics = 40
for i in range(n_harmonics):
    f = 2*i + 1 # Move through odd numbers
    y = y + np.sin(2*np.pi*f*t)/f
plt.plot(y)
```
